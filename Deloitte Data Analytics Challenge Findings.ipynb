```python

from matplotlib.lines import Line2D

#file path
file_path = "/Users/IshikaKshetrapal/Documents/transaction.xlsx"
df = pd.read_excel(file_path)

# Filter transactions for only grocers
merchants = ['Whole Foods Market', 'Aldi', 'Safeway', 'Target']
filtered_df = df[df['Merchant'].isin(merchants)]

# I wanted to do a pastel color plot
pastel_palette = sns.color_palette("pastel")

# creation of the violion plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=filtered_df, x='Merchant', y='Amount', palette=pastel_palette)

# adding an average spending amount for each grocer 
for i, merchant in enumerate(merchants):
    avg_amount = filtered_df[filtered_df['Merchant'] == merchant]['Amount'].mean()
    plt.axhline(avg_amount, color=pastel_palette[i], linestyle='--', linewidth=1)

# legend creation
legend_handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor=pastel_palette[i], markersize=10, label=f'Avg {merchant} Amount: ${filtered_df[filtered_df["Merchant"] == merchant]["Amount"].mean():.2f}') for i, merchant in enumerate(merchants)]
plt.legend(handles=legend_handles, loc='upper left')

# formatting title so its easy to read(no overlapping elements)
plt.suptitle('Transaction Amounts for Different Grocery Merchants', y=1.05)
plt.xlabel('Merchant')
plt.ylabel('Transaction Amount')
plt.tight_layout()

plt.show()

```
